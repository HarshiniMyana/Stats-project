```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
df = pd.read_csv('/content/Sales Data.csv')



# Display the first few rows and summary statistics
print(df.head())

```

```python

# Handle missing values
df.isna().sum() #checks if there any missing values
df.drop_duplicates(inplace=True) #removes dupicates or drop duplicate rows
df=pd.read_csv('/content/Sales Data.csv', index_col = 0)
df.head()


# Convert data types if necessary
df.dtypes # shows the datatype for each column
# column product, order date, purchase Address, city are set as dtype :object
df['Product'] = df['Product'].astype('category') # converted data type object to category
df['Order Date'] = pd.to_datetime(df['Order Date'])# converted data type object to datetime
df['Purchase Address'] = df['Purchase Address'].astype('category')#converted data type to category
df['City'] = df['City'].astype('category')#converted data type to category
df.dtypes

```

```python
#Handling missing values
df.isna().sum()
df['Price Each'] = df['Price Each'].fillna(df['Price Each'].mean())# Removes rows with missing values
df['Quantity Ordered'] = df['Quantity Ordered'].fillna(df['Quantity Ordered'].mean())# Removes rows with missing values
df['Order Date'] = df['Order Date'].fillna(df['Order Date'].mean())# Removes rows with missing values
df['Month'] = df['Month'].fillna(df['Month'].mean())# Removes rows with missing values
df['Hour'] = df['Hour'].fillna(df['Hour'].mean())# Removes rows with missing values
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())# Removes rows with missing values
df['City'] = df['City'].fillna(df['City'].mode()[0])# Removes rows with missing values
df['Purchase Address'] = df['Purchase Address'].fillna(df['Purchase Address'].mode()[0])# Removes rows with missing values



```

```python

from scipy.stats import ttest_ind, chi2_contingency

# Example: T-test for sales in different regions
import pandas as pd
from scipy import stats

# Select the 'Sales' column for the t-test
sales_data = df['Sales']

# Perform the one-sample t-test on the 'Sales' column
stats.ttest_1samp(sales_data, 185950)
# Get unique cities
cities = df['City'].unique()

# Iterate over pairs of cities
for i in range(len(cities)):
    for j in range(i + 1, len(cities)):
        city_1 = cities[i]
        city_2 = cities[j]

        # Filter sales data for each city
        city_1_sales = df[df['City'] == city_1]['Sales']
        city_2_sales = df[df['City'] == city_2]['Sales']

        # Perform T-test
        t_stat, p_value = stats.ttest_ind(city_1_sales, city_2_sales, nan_policy='omit')

        # Print results
        print(f"Comparing {city_1} and {city_2}:")
        print(f"T-statistic: {t_stat}, P-value: {p_value}")

        # Interpretation
        if p_value < 0.3:
            print(f"There is a significant difference in sales between {city_1} and {city_2}.")
        else:
            print(f"There is no significant difference in sales between {city_1} and {city_2}.")
        print("-" * 40)







# Example: Chi-square test for independence between Ship Mode and Customer Segment
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['City'], df['Product'])

# Perform the chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n", expected)

# Interpret the results
alpha = 0.3  # Significance level
if p_value < alpha:
    print("There is a significant association between City and Product.")
else:
    print("There is no significant association between City and Product.")



```

```python
#plotting sales trends over time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1: Aggregating Sales by Month
df['Month'] = df['Order Date'].dt.to_period('M')  # Aggregates by month (you can use 'Y' for year, 'D' for day)
sales_by_month = df.groupby('Month')['Sales'].sum()

# Plotting Sales Trends by Month
plt.figure(figsize=(10, 6))
sns.lineplot(x=sales_by_month.index.astype(str), y=sales_by_month.values, marker='o', color='b')
plt.title('Sales Trends Over Time by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 2: Aggregating Sales by Day
df['Day'] = df['Order Date'].dt.date  # Aggregates by day
sales_by_day = df.groupby('Day')['Sales'].sum()

# Plotting Sales Trends by Day
plt.figure(figsize=(12, 6))
sns.lineplot(x=sales_by_day.index, y=sales_by_day.values, marker='o', color='g')
plt.title('Sales Trends Over Time by Day')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 3: Aggregating Sales by Year
df['Year'] = df['Order Date'].dt.year  # Aggregates by year
sales_by_year = df.groupby('Year')['Sales'].sum()

# Plotting Sales Trends by Year
plt.figure(figsize=(10, 6))
sns.lineplot(x=sales_by_year.index, y=sales_by_year.values, marker='o', color='r')
plt.title('Sales Trends Over Time by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
#plotting profit trends over time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assumed df has 'Order Date' and 'Sales' columns

# 1. Estimate Profit using a Profit Margin
profit_margin = 0.2  # Replacing  estimated profit margin (e.g., 0.2 for 20%)
df['Profit'] = df['Sales'] * profit_margin



# 3. Aggregate by Month
df['Month'] = df['Order Date'].dt.to_period('M')  # Aggregating by Month
profit_by_month = df.groupby('Month')['Profit'].sum()

# 4. Plotting Profit Trends by Month
plt.figure(figsize=(10, 5))
sns.lineplot(x=profit_by_month.index.astype(str), y=profit_by_month.values, marker='o', color='b')
plt.title('Estimated Profit Trends Over Time by Month')
plt.xlabel('Month')
plt.ylabel('Total Estimated Profit')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
#Identifying top-performing product categories based on total sales and profit
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assumed df has 'Product', 'Sales', and 'Profit' columns

# 1. Calculate Total Sales and Profit by Product Category
product_performance = df.groupby('Product').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')  # Assuming 'Profit' column was calculated previously
)

#  Sort by Total Sales and Profit
top_sales = product_performance.sort_values('Total_Sales', ascending=False)
top_profit = product_performance.sort_values('Total_Profit', ascending=False)

#  Display Top Performers
print("Top Performing Product Categories by Sales:\n", top_sales.head())
print("\nTop Performing Product Categories by Profit:\n", top_profit.head())

#2.Plotting Total Sales by Product Category
plt.figure(figsize=(12, 6))
sns.barplot(x=top_sales.index, y=top_sales['Total_Sales'], color='skyblue')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# 3.Plotting Total Profit by Product Category
plt.figure(figsize=(12, 6))
sns.barplot(x=top_profit.index, y=top_profit['Total_Profit'], color='lightgreen')
plt.title('Total Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
